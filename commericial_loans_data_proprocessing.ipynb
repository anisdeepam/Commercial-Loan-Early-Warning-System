```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
```

```python
# Load dataset
df = pd.read_csv('synthetic_commercial_loans_final.csv')
```

```python
print('Original Shape:', df.shape)
```

```python
print('Missing Values:', df.isnull().sum().sum())
```

```python
#Encoding
categorical_cols = ['Loan_Purpose', 'Industry']
df_encoded = pd.get_dummies(df, columns=categorical_cols, drop_first=True, dtype=int)
```

```python
# Scale numeric features
numeric_cols = df_encoded.select_dtypes(include=[np.number]).columns.drop(['Default'], errors='ignore').tolist()
scaler = StandardScaler()
df_scaled = df_encoded.copy()
df_scaled[numeric_cols] = scaler.fit_transform(df_encoded[numeric_cols])
```

```python
df_encoded.to_csv('preprocessed_commercial_loans.csv', index=False)
df_scaled.to_csv('preprocessed_scaled_commercial_loans.csv', index=False)
```
